# first make a list of prime numbers less than 10,000

```python
import sympy
primeNumbers=list()
```

```python
for number in range(2,10001):
        if sympy.isprime(number):
            primeNumbers.append(number)
    
```

displaying Prime numbers

```python
primeNumbers
```

# create a sum array

```python
SumList = list()
sum =0;
for number in primeNumbers:
    sum +=number 
    SumList.append(sum)
SumList
```

create list named Result to add to it the consecutive sum 
checking the consecutive sum starting from 2

```python
Result = list()
for s in SumList:
    if s==2: continue
    if sympy.isprime(s):
        if s<=100000:
            Result.append(s)
Result    
```

checking the consecutive prime numbers but not from start and then sorting the list

```python
i=0
j=0
for sum in SumList:
    i+=1
    j=0
    newsum =sum
    for prime in primeNumbers:
        j+=1
        if j>=i-1:
            break
        newsum-=prime
        if sympy.isprime(newsum):
            if (newsum)<=100000:
                Result.append(newsum)       
Result.sort()     
Result
    
    
```

remove the repeated elements and display the final result

```python
FinalResult = list()
prev= 2
for number in Result:
    if prev==number:
        continue
    else :
        FinalResult.append(number)
    prev = number
FinalResult
```
